## Pie chart 3
Algo mas complejo

Caracteristicas

    .Se indican los colores

    .Le digo que empiece a plotear desde un angulo de 90º

    .Radio del pie chart 1.0

```python
y=['futbol','baloncesto','gimnasio','TV']
x=[4,2,4,1]

plt.title("horas semanales de tiempo libre")
plt.pie(x, labels=y,
       colors=["r","y","g","b"],
       startangle=90,
       radius=1.0)
plt.show()
```

## Pie chart 4
autopct le indica que escriba el numero dentro( % )

```python
y=['futbol','baloncesto','gimnasio','TV']
x=[4,2,4,1]

plt.title("horas semanales de tiempo libre")
plt.pie(x, labels=y,
       colors=["r","y","g","b"],
       startangle=90,
       radius=1.1,
       autopct= "%1.2f%%")
plt.show()
```

# PRINCIPALES TIPOS DE GRAFICOS[4]: VARIOS PLOTS
## Varios plots

```python
import numpy as np
plt.figure(figsize=(12,8))
# lo metes todo en una figura para que no se te quede pequeño..
x=np.arange(-10,10,0.01)
# simplemente le das el rango (-10,10) de 0.01 a 0.01

# simplemente le dices los subplots
# (numero de filas, numero de columnas, numero de grafico)
plt.subplot(2,2,1)
y1=x
plt.plot(x,y1)
plt.title("y=x")

plt.subplot(2,2,2)
y2=x**2
plt.plot(x,y2)
plt.title("y es x al cuadrado")

plt.subplot(2,2,3)
y3=x**3
plt.plot(x,y3)
plt.title("y es x al cubo")

plt.subplot(2,2,4)
y4=x**4
plt.plot(x,y4)
plt.title("y es x a la cuarta")

plt.show()
```

## Pregunta 2 – (hasta 1 punto)

### ¿ Qué son Q1, Q2, Q3 ?

Partiendo del ejemplo para explicarlo:
    
Para saber a que corresponde Q1,Q2,Q3, una vez que se ha hecho el calculo, se coge el primer valor que sobrepasa en la acumulada.

Q1= que es el percentil 25,es decir, P25%, hasta que valor esta el 25% de los datos. Para calcular: **N/4**

   15/4= 3,75 ----  **Q1= 57**   

Q2= equivale a la mediana. Para calcular: **2N/4 == N/2**

   15/2= 7,5 ----  **Q2= 65** 

Q3= P75%, hasta el valor que esta el 75% de los datos. Para calcular:  **3N/4**

   3x15/4= 11,25 ---- **Q3= 71**

## Pregunta 3 – (hasta 1 punto)

### ¿Qué son los "outliers" ? ¿ son mostrados en estos diagramas de cajas?

Un valor atípico (en inglés outlier) es una observación que es numéricamente distante del resto de los datos. En ocasiones
puede deberse a errores en la recogida de datos. Estos deben ser estudiados con detenimiento con el fin de decidir
si deben o no ser eliminados del análisis estadístico.

Siguiendo con el ejemplo

El rango intercuartílico = **RIQ= Q3 - Q1** = 71 - 57 =    **14=RIQ**

**f1= Q1 - 1,5 RIQ** =  57 - 1,5 X 14 =  **36=f1**  son los valores atipicos leves

**f3= Q1 + 1,5 RIQ** =  57 + 1,5 X 14 =  **92=f3**  son los valores atipicos leves

Por lo que el valor atipico seria 96, como se ve en la figura del ejemplo se muestra con un circulo.
