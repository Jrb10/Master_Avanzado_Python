```python
import dash
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input,Output
import plotly.express as px
import quandl
import nasdaqdatalink




app = JupyterDash(__name__)

app.layout=html.Div(children=[
    html.H1(children="APLICACION CON DASH PARA MERCADOS FINANCIEROS"),
    html.H3("ELIGE LA FECHA Y LA EMPRESA "),
    dcc.DatePickerRange(id='date_range',
                       min_date_allowed="2014-01-01",
                       max_date_allowed="2018-01-01",
                       start_date="2014-01-01",
                       end_date="2018-01-01"),
    
    
    dcc.Dropdown(['ABC', 'BBCN','CACB','DLB','ENDP','FDO','GTY','ICPT'], 
                  'ABC', id='empresa'),
    dcc.Graph( id='grafico')
])
    

@app.callback(

    Output('grafico', 'figure'),
    [
        Input('date_range', 'start_date'),
        Input('date_range', 'end_date'),
        Input('empresa', 'value')
    ]
    
    
)

def update_output(start_date,end_date,value):    
    
    if value=='ABC':
        valueC='WIKI/'+value+'.4'
        df=quandl.get(valueC,start_date=start_date, end_date=end_date)
        figura=px.line(df, title=f"EMPRESA {value} ")
        return figura
    else:
        
        valueC='WIKI/'+value+'.4'
        df=quandl.get(valueC, start_date=start_date, end_date=end_date)
        figura=px.line(df, title=f"COTITACION {value} ")
        return figura

if __name__== "__main__":
    app.run_server(mode="external")
```
